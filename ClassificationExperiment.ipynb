```python
import numpy as np
from sklearn.datasets import load_svmlight_file
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import random


# define model function
def model_function(x, omega):
    return np.dot(x, omega)


# define model function with threshold
def model_function_t(x, threshold):
    if model_function(x) > threshold:
        return 1
    else:
        return -1


# define calculate gradient function
def calculate_gradient(given_x, given_y, omega):
    xi = 1 - given_y * np.dot(given_x, omega)
    symbols = np.maximum(0, xi / np.abs(xi))

    y = given_y * symbols
    d_w = -np.dot(given_x.T, y) / given_y.size
    return C * d_w + omega


# define loss function
def loss_function(given_x, given_y, omega):
    xi = 1 - given_y * np.dot(given_x, omega)
    A = sum(np.maximum(0, xi))
    loss = 0.5 * np.dot(omega, omega) + (A * C) / given_y.size
    return loss


# load data set
X_train, y_train = load_svmlight_file("a9a")
X_train = X_train.toarray()
X_train = np.column_stack((X_train, np.ones(X_train.shape[0])))

# load validation date set
X_validation, y_validation = load_svmlight_file("a9a.t")
X_validation = X_validation.toarray()
X_validation = np.column_stack((X_validation, np.zeros(X_validation.shape[0])))
X_validation = np.column_stack((X_validation, np.ones(X_validation.shape[0])))


# define learning rate
eta = 0.001

# define c
C = 0.9

# define mu (NAG)
mu = 0.9

# define decay rate (AdaDelta) (RMSProp)
gamma = 0.9

# define beta_1 beta_2 (Adam)
beta_1 = 0.9
beta_2 = 0.999

# define epsilon
epsilon = 0.00000001

# define iteration number
iteration_num = 1000

# initial loss list
SGD_loss = []
NAG_loss = []
RMSProp_loss = []
AdaDelta_loss = []
Adam_loss = []

# initial AdaDelta parameters
Ada_E_Theta = 0
Ada_E_g = 0
Ada_delta_theta = np.ones(124) / 100

# initial RMSProp parameters
RMS_E_Theta = 0
RMS_E_g = 0
RMS_delta_theta = np.ones(124) / 100

# initial Adam parameters
v_t = 0
m_t = np.zeros(124)
Adam_delta_theta = np.zeros(124)

# initial Adam momentum
momentum = np.zeros(124)

SGD_gradient = np.zeros(124)
NAG_gradient = np.zeros(124)
AdaDelta_gradient = np.zeros(124)
RMSProp_gradient = np.zeros(124)
Adam_gradient = np.zeros(124)

# parameters initialize - zeros
SGD_omega = np.zeros(124)
NAG_omega = np.zeros(124)
RMSProp_omega = np.zeros(124)
AdaDelta_omega = np.zeros(124)
Adam_omega = np.zeros(124)

# define batch number
batch_number = int(y_train.size / 7)

# gradient decent
for i in range(iteration_num):
    # print current iteration number
    print("\rProducing (" + str(i + 1) + "/" + str(iteration_num) + ")", end="")

    # select samples' index
    index = random.randint(0, y_train.size - batch_number)

    # calculate gradient
    SGD_gradient = calculate_gradient(X_train[i:i + batch_number], y_train[i:i + batch_number], SGD_omega)
    NAG_gradient = calculate_gradient(X_train[i:i + batch_number], y_train[i:i + batch_number], NAG_omega + mu * momentum)
    AdaDelta_gradient = calculate_gradient(X_train[i:i + batch_number], y_train[i:i + batch_number], AdaDelta_omega)
    RMSProp_gradient = calculate_gradient(X_train[i:i + batch_number], y_train[i:i + batch_number], RMSProp_omega)
    Adam_gradient = calculate_gradient(X_train[i:i + batch_number], y_train[i:i + batch_number], Adam_omega)

    # calculate new momentum (NAG)
    momentum = mu * momentum - eta * NAG_gradient

    # calculate E_Theta and E -g (AdaDelta)
    Ada_E_Theta = gamma * Ada_E_Theta + (1 - gamma) * np.dot(Ada_delta_theta, Ada_delta_theta)
    Ada_E_g = gamma * Ada_E_g + (1 - gamma) * np.dot(AdaDelta_gradient, AdaDelta_gradient)

    # calculate delta theta (AdaDelta)
    Ada_delta_theta = (-np.sqrt(Ada_E_Theta + epsilon) / np.sqrt(Ada_E_g + epsilon)) * AdaDelta_gradient

    # calculate E_Theta and E -g (RMSProp)
    RMS_E_Theta = gamma * RMS_E_Theta + (1 - gamma) * np.dot(RMS_delta_theta, RMS_delta_theta)
    RMS_E_g = gamma * RMS_E_g + (1 - gamma) * np.dot(RMSProp_gradient, RMSProp_gradient)

    # calculate delta theta (RMSProp)
    RMS_delta_theta = (- eta / np.sqrt(RMS_E_g + epsilon)) * RMSProp_gradient

    # calculate mt and vt (Adam)
    m_t = beta_1 * m_t + (1 - beta_1) * Adam_gradient
    v_t = beta_2 * v_t + (1 - beta_2) * np.dot(Adam_gradient, Adam_gradient)

    # revises mt and vt (Adam)
    m_t_r = m_t / (1 - np.power(beta_1, i + 1))
    v_t_r = v_t / (1 - np.power(beta_2, i + 1))

    # calculate delta theta (Adam)
    Adam_delta_theta = (- eta / (np.sqrt(v_t_r) + epsilon)) * m_t_r

    # update parameters
    SGD_omega = SGD_omega - eta * SGD_gradient
    NAG_omega = NAG_omega + momentum
    AdaDelta_omega = AdaDelta_omega + Ada_delta_theta
    RMSProp_omega = RMSProp_omega + RMS_delta_theta
    Adam_omega = Adam_omega + Adam_delta_theta

    # calculate loss
    SGD_loss.append(loss_function(X_validation, y_validation, SGD_omega))
    NAG_loss.append(loss_function(X_validation, y_validation, NAG_omega))
    AdaDelta_loss.append(loss_function(X_validation, y_validation, AdaDelta_omega))
    RMSProp_loss.append(loss_function(X_validation, y_validation, RMSProp_omega))
    Adam_loss.append(loss_function(X_validation, y_validation, Adam_omega))

# plot loss changes with iteration
# yellow line for validation data loss
plt.plot(range(len(SGD_loss)), SGD_loss, label="SGD")
plt.plot(range(len(NAG_loss)), NAG_loss, label="NAG")
plt.plot(range(len(AdaDelta_loss)), AdaDelta_loss, label="AdaDelta")
plt.plot(range(len(RMSProp_loss)), RMSProp_loss, label="RMSProp")
plt.plot(range(len(Adam_loss)), Adam_loss, label="Adam")
plt.legend()
plt.yscale("log")

plt.xlabel("Iteration number")
plt.ylabel("Loss")
plt.title("Linear Classification and gradient decent")
plt.show()
```
